````python
import pandas, warnings

import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

plt.style.use("seaborn-whitegrid")
````

````python
reader = lambda url, **multiargs : pandas.read_table(url, sep=",", **multiargs)

url = "okajakpril.csv"

origin = reader(url, parse_dates=["tap_in_time", "tap_out_time"])
````

### ```Helper```

````python
def barplot(result, **barlimit):
  if "ylabel" not in barlimit.keys():
    barlimit["ylabel"] = "Total Data"
  bar = result.plot.bar(rot=0, **barlimit)
  bar.bar_label(bar.containers[0], fmt="%1.1f");

colors = ["#BC9F8B", "#B5CFB7", "#C5705D", "#ACE1AF", "#B3C8CF"]
````

### ```General on April 2023```

Understand Profit, Trip in April 2023 Without Further Analyze

````python
total_profit = origin.profit.sum()

total_trip = origin.profit.count()

total_profit
````

#### ```Profit Distribution```

````python
result = origin.profit.value_counts()

barplot(result, xlabel="Payment", title="Total Payment", color=colors)
````

````python
import numpy as np

nul = origin[origin["profit"] == 0]

nul["jak"] = np.where(nul["corridor_id"].str.startswith("JAK"), 1, 0)

nul.iloc[:5, 16:]
````

````python
jato = nul["jak"].value_counts()

title = "Jaklingko in 0 Table"

barplot(jato, xlabel="Jaklingko", title=title, color=colors)
````

````python
result = origin.pay_card_sex.value_counts()

title = "Total Customer Gender"

barplot(result, xlabel="Gender", title=title, color=colors)
````

````python
origin["age"] = 2023 - origin.pay_card_birth_date

title = "Total Age Distribution"

plot = origin["age"].plot.hist(title=title, bins=10, xlabel="Age", color=colors[3])

plot.bar_label(plot.containers[0]);
````

### ```Card Service```

````python
result = origin.pay_card_bank.value_counts()

title = "Total Customer Pay Card"

provider = ["DKI", "E Money", "Brizzi", "Flazz", "Online", "BNI"]

plot = result.plot.bar(rot=0, title=title, color=colors, xlabel="Card")

plot.set_xticklabels(provider, rotation=0)

plot.set_ylabel("Total Data")

plot.bar_label(plot.containers[0], fmt="%1.1f");
````

### ```Corridor```

````python
corr = origin.corridor.value_counts().iloc[:4]

barplot(corr, xlabel="Corridor", title="Total Corridor", color=colors, figsize=(11, 5))
````

````python
corr = origin.groupby("corridor")["selisih"].mean()

corr = corr.sort_values(ascending=False).iloc[:4]

title = "Average Time Trip in Corridor"

ylabel = "Average (Min)"

barplot(corr, xlabel="Corridor", title=title, color=colors, figsize=(11, 5), ylabel=ylabel)
````

````python
corr = origin.groupby("corridor")["selisih"].mean()

corr = corr.sort_values().iloc[:4]

title = "Average Time Trip in Corridor"

ylabel = "Average (Min)"

barplot(corr, xlabel="Corridor", title=title, color=colors, figsize=(11, 5), ylabel=ylabel)
````

### ```Direction```

````python
corr = origin.direction.value_counts()

barplot(corr, xlabel="Direction", title="Total Direction", color=colors)
````

````python
result = origin.groupby("direction")["profit"].sum()

title = "Profit over Direction"

barplot(result, xlabel="Direction", title=title, color=colors, ylabel="Total Profit")
````

````python
result = origin.groupby("direction")["selisih"].mean()

title = "Trip Time v Direction"

barplot(result, xlabel="Direction", title=title, color=colors, ylabel="Average Time Trip")
````

### ```Hour```

````python
hourly = origin.groupby("in_hour")["profit"].count()

hourly = hourly.sort_values(ascending=False).iloc[:10]

barplot(hourly, title="Total Jam Trip", xlabel="Jam")
````

````python
hourly = origin.groupby("in_hour")["selisih"].mean()

hourly = hourly.sort_values().iloc[:10]

barplot(hourly, title="Average Time on Hour", xlabel="Jam", ylabel="Average Time Trip")
````

### ```Hari```

````python
daily = origin.groupby("in_day")["profit"].count()

daily = daily.sort_values(ascending=False)

barplot(daily, title="Total Harian Trip", xlabel="Hari")
````
